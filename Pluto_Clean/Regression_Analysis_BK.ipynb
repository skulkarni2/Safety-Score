```python
RA_BK.rename(columns={'q-score': 'q', 'Unnamed: 0': 'No_Name'}, inplace=True)
```

```python
plt.scatter(RA_BK['q'],RA_BK['YearBuilt'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ YearBuilt', data = RA_BK).fit()
```

```python
print lm.params
print lm.rsquared
print lm.summary()
```

```python
RA_BK['Building_Age'] = 2016 - RA_BK['YearBuilt']
RA_BK
```

```python
RA_BK['Building_Age'].dtype
```

```python
RA_BK[RA_BK.Building_Age == 2016]
```

```python
for i in range(len(RA_BK)):
    if RA_BK.loc[i,'Building_Age'] == 2016:
        RA_BK.loc[i,'Building_Age'] = 1
RA_BK
```

```python
plt.scatter(RA_BK['q'],RA_BK['Building_Age'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ Building_Age', data = RA_BK).fit()
print lm.params
print lm.summary()
```
