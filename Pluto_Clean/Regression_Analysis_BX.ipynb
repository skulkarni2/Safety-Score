```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pylab as plt
%matplotlib inline
```

```python
RA_BX = pd.read_csv('/Users/achillessaxby/gittest_achilles/Safety_Score-Pluto_Cleaning/Data/ShapeFiles/BX_Geometry_Joined.csv')
RA_BX.shape
```

```python
RA_BX.rename(columns={'q-score': 'q', 'Unnamed: 0': 'No_Name'}, inplace=True)
```

```python
plt.scatter(RA_BX['q'],RA_BX['YearBuilt'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ YearBuilt', data = RA_BX).fit()
```

```python
print lm.params
print lm.rsquared
print lm.summary()
```

```python
RA_BX['Building_Age'] = 2016 - RA_BX['YearBuilt']
RA_BX
```

```python
RA_BX['Building_Age'].dtype
```

```python
RA_BX[RA_BX.Building_Age == 2016]
```

```python
for i in range(len(RA_BX)):
    if RA_BX.loc[i,'Building_Age'] == 2016:
        RA_BX.loc[i,'Building_Age'] = 1
RA_BX
```

```python
plt.scatter(RA_BX['q'],RA_BX['Building_Age'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ Building_Age', data = RA_BX).fit()
print lm.params
print lm.summary()
```
