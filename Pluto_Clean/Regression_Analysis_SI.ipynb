```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pylab as plt
%matplotlib inline
```

```python
RA_SI.rename(columns={'q-score': 'q', 'Unnamed: 0': 'No_Name'}, inplace=True)
```

```python
plt.scatter(RA_SI['q'],RA_SI['YearBuilt'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ YearBuilt', data = RA_SI).fit()
```

```python
print lm.params
print lm.rsquared
print lm.summary()
```

```python
RA_SI['Building_Age'] = 2016 - RA_SI['YearBuilt']
RA_SI
```

```python
RA_SI['Building_Age'].dtype
```

```python
RA_SI[RA_SI.Building_Age == 2016]
```

```python
for i in range(len(RA_SI)):
    if RA_SI.loc[i,'Building_Age'] == 2016:
        RA_SI.loc[i,'Building_Age'] = 1
RA_SI
```

```python
plt.scatter(RA_SI['q'],RA_SI['Building_Age'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ Building_Age', data = RA_SI).fit()
print lm.params
print lm.summary()
```
