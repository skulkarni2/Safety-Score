```python
RA_MN.rename(columns={'q-score': 'q', 'Unnamed: 0': 'No_Name'}, inplace=True)
```

```python
plt.scatter(RA_MN['q'],RA_MN['YearBuilt'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ YearBuilt', data = RA_MN).fit()
```

```python
print lm.params
print lm.rsquared
print lm.summary()
```

```python
RA_MN['Building_Age'] = 2016 - RA_MN['YearBuilt']
RA_MN
```

```python
RA_MN['Building_Age'].dtype
```

```python
RA_MN[RA_MN.Building_Age == 2016]
```

```python
for i in range(len(RA_MN)):
    if RA_MN.loc[i,'Building_Age'] == 2016:
        RA_MN.loc[i,'Building_Age'] = 1
RA_MN
```

```python
plt.scatter(RA_MN['q'],RA_MN['Building_Age'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ Building_Age', data = RA_MN).fit()
print lm.params
print lm.summary()
```
