```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pylab as plt
%matplotlib inline
```

```python
RA_QN = pd.read_csv('/Users/achillessaxby/gittest_achilles/Safety_Score-Pluto_Cleaning/Data/ShapeFiles/QN_Geometry_Joined.csv')
RA_QN.shape
```

```python
RA_QN.rename(columns={'q-score': 'q', 'Unnamed: 0': 'No_Name'}, inplace=True)
```

```python
plt.scatter(RA_QN['q'],RA_QN['YearBuilt'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ YearBuilt', data = RA_QN).fit()
```

```python
print lm.params
print lm.rsquared
print lm.summary()
```

```python
RA_QN['Building_Age'] = 2016 - RA_QN['YearBuilt']
RA_QN
```

```python
RA_QN['Building_Age'].dtype
```

```python
RA_QN[RA_QN.Building_Age == 2016]
```

```python
for i in range(len(RA_QN)):
    if RA_QN.loc[i,'Building_Age'] == 2016:
        RA_QN.loc[i,'Building_Age'] = 1
RA_QN
```

```python
plt.scatter(RA_QN['q'],RA_QN['Building_Age'])
plt.show()
```

```python
lm = smf.ols(formula='q ~ Building_Age', data = RA_QN).fit()
print lm.params
print lm.summary()
```
